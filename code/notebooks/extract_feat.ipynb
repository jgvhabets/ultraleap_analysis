Function that creates dataframe with all features per task

```python
def get_scores(sub, cond, cam, task, side, block):

    """
    
    """

    read_scores = pd.read_excel(
        os.path.join(
        find_paths.find_onedrive_path('patientdata'),
        f'scores_JB_JH_JR.xlsx'),
        usecols='A:I'
        )

    read_scores.set_index('sub_cond_cam', inplace = True)

    if side == 'left': side='lh'
    elif side == 'right': side='rh'
    
    # read scores for all blocks of a subject in the same cond, cam per side
    ext_scores = read_scores.loc[f'{sub}_{cond}_{cam}'][f'{task}_{side}']

    if type(ext_scores) != float:
        ls_int_sc = [int(s) for s in ext_scores if s in ['0', '1', '2', '3', '4']]
        
        if block == 'b1':
            score = ls_int_sc[0]
        elif block == 'b2':
            score = ls_int_sc[1]
        elif block == 'b3':
            score = ls_int_sc[2]
        else:
            print(f'no scores for block {block} or block does not exist')
        return score
```

```python
def feat_over_tap(sub, task, cond, cam, block, side, feat_name,):
    # C:\Users\madal\Thesis\ultraleap_analysis\data\ul001\ft\m1\distances
    block_path = pd.read_csv(os.path.join(repo_path, 'data', sub, task, cond, 'distances', f'{block}_{sub}_{cond}_{cam}_{task}_{side}.csv'))
    # get_time = block_path.time.to_list()

    if task == 'oc':
        _, idx_min = hp.find_min_max(np.array(block_path['mean']), cam=cam)
        ls_time,_,_ = feat_ext.tap_times(block_path, 'mean', idx_min)
    elif task == 'ft':
        _, idx_min = hp.find_min_max(np.array(block_path['dist']), cam=cam)
        ls_time,_,_ = feat_ext.tap_times(block_path, 'dist', idx_min)

    # C:\Users\madal\Thesis\ultraleap_analysis\features\ul001\ft\m1\features_tap
    get_feat = json.load(open(os.path.join(repo_path, 'features', sub, task, cond, 'features_tap', f'{block}_{sub}_{cond}_{cam}_{task}_{side}.json'))) 
    n_taps = np.arange(0,len(get_feat[feat_name]))
    
    times = []
    for t in ls_time:
        times.append(t[0])

    # print(ls_time)
    # plt.plot(times[:-1], get_feat[feat_name], marker = 'o')#, ms = 2)
    plt.plot(n_taps, get_feat[feat_name], marker = 'o', ms = 2, color = 'C3')

    if task == 'ft':
        plt.xlabel('# taps')

    if task == 'oc':
        plt.xlabel('# oc events')

    if 'dist' in feat_name:
        plt.ylabel('distance (m)')
        plt.title(feat_name)
        # plt.legend([legend_name])
    elif 'vel' in feat_name:
        plt.ylabel('speed (m/s)')
        plt.title(feat_name)
        # plt.legend([legend_name])
    elif feat_name == 'rms':
        plt.ylabel('distance (m)')
        plt.title(feat_name)


    return #get_feat[feat_name]
```

```python
def plot_feat_score(sub, task, cam, cond, block, score, feat): 

    if cam == 'dt' or cam == 'vr':
        feat_open = open(os.path.join(repo_path,'features', sub, task, cond, 'features_block', f'{block}_{sub}_{cond}_{cam}_{task}_right.json'))
        feat_load = json.load(feat_open)

        if cond == 'm1s0': marker_type = 'o'
        elif cond == 'm1s1': marker_type = '^'
        else: 
            marker_type = '^'
        
        plt.scatter(score, feat_load[feat], marker = marker_type, color = 'k')
        plt.ylabel(feat)
        plt.xlabel('score')
        plt.xticks([0, 1, 2, 3])
        plt.xlim(-0.5,3.5)
        # plt.yticks([y for y in np.arange(0, 0.10, 0.005)])
        plt.title('UPDRS sub-scores for patients in different conditions')
        plt.tight_layout()

    return
```
